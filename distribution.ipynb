1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times?

-Each die roll has:

Probability of rolling a 3 = 1/6
Number of trials (n) = 30
Number of successes (x) = 5
Excel Formula
=BINOM.DIST(5,30,1/6,FALSE)
Result

Probability ≈ 0.184 (18.4%)

6. Why is Uniform Distribution a good model for a fair die?

A fair die has six outcomes:

1, 2, 3, 4, 5, 6

Each outcome has the same probability.

P=
6
1
	​


Since every outcome is equally likely, the die follows a Discrete Uniform Distribution.

Properties
Equal probability for each face.
No outcome is favored.
Total probability equals 1.

Therefore, the uniform distribution accurately models a fair die.

7. Excel: Probability of at least 8 successes in 15 trials (p = 0.5)

-Given

n = 15
p = 0.5
Need P(X≥8)
Excel Formula
=1-BINOM.DIST(7,15,0.5,TRUE)
Result

Probability ≈ 0.5001 (50.01%)

8. How does Log Transformation help?
Definition

-A log transformation replaces each value with its logarithm.

Examples:

LN(x)
LOG10(x)
Benefits
Reduces skewness.
Stabilizes variance.
Compresses large values.
Makes data closer to a normal distribution.
Improves regression and machine learning performance.

Applications

Conclusion: Log transformation reduces the impact of extreme values, stabilizes variance, and often makes data more normally distributed, improving the reliability of statistical analyses.
