Question 4

The population of a school is 60% boys and 40% girls. If equal representation of both genders is required, which sampling method should be used? Why?

Answer:

Stratified Sampling should be used because the population is divided into different groups (boys and girls), and this method ensures both groups are properly represented in the sample.

Final Answer: Stratified Sampling

Question 5

The average height of 1000 students is 160 cm. A sample of 100 students has an average height of 158 cm. Find the sampling error.

Answer:

Sampling Error

=Sample Mean−Population Mean
=158−160=−2 cm

Final Answer: −2 cm (Magnitude = 2 cm)
