Q4: Median in Real Life

Prices: [$8,000, $9,500, $10,200, $11,000, $50,000]

Since the values are already sorted,

Median = Middle Value = $10,200

Answer:
The median is better than the mean because the $50,000 car is an outlier that would increase the average significantly. The median better represents the typical used-car price.
