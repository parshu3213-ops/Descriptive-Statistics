Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

Answer:
The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship between variables.

Importance:

It is the starting assumption in hypothesis testing.
It helps determine whether the sample data provides enough evidence to reject the claim.
Statistical tests are designed to test the validity of H₀.

Example:

H₀: The average salary of employees is ₹50,000.
H₁: The average salary is not ₹50,000.
Question 2: What does the significance level (α) represent in hypothesis testing?

Answer:
The significance level (α) is the probability of making a Type I error (rejecting a true null hypothesis).

Common values:

α = 0.05 (5%)
α = 0.01 (1%)

If p-value < α, reject H₀.
If p-value ≥ α, fail to reject H₀.

Question 3: Differentiate between Type I and Type II errors.
Type I Error	Type II Error
Rejecting a true H₀	Failing to reject a false H₀
False Positive	False Negative
Probability = α	Probability = β

Example:
Medical Test:

Type I: Healthy person diagnosed as sick.
Type II: Sick person diagnosed as healthy.
Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.
One-Tailed Test	Two-Tailed Test
Tests for increase or decrease	Tests for any difference
One rejection region	Two rejection regions
One-tailed Example

H₀: μ = 100

H₁: μ > 100

(Testing whether production increased.)

Two-tailed Example

H₀: μ = 100

H₁: μ ≠ 100

(Testing whether production changed.)

Question 6: When should you use a Z-test instead of a t-test?

Use a Z-test when:

Population standard deviation (σ) is known.
Sample size is large (n ≥ 30).
Data is approximately normally distributed.

Use a t-test when:

Population standard deviation is unknown.
Sample size is small (n < 30).
